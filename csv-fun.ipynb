Working with Databases in Julia

```julia
#We are going to use SQL Lite
#Pkg.add("SQLite")
```

```julia
#DDL - Create, Alter and drop - SQLite.DB
#DML - Insert, Update and Delete - SQLite.execute
#DQL - Select - SQLite.Query
```

```julia
using SQLite
db = SQLite.DB("Movies")
```

```julia
SQLite.execute(db, "CREATE TABLE IF NOT EXISTS movies(movie_id REAL, movie_name TEXT, location TEXT)")
```

```julia
SQLite.tables(db)
```

```julia
SQLite.execute(db,"INSERT INTO movies (movie_id,movie_name,location) VALUES(1, 'Donnie Darko', 'USA')")
```

```julia
#We would previously use DataFrame(SQLite.execute()) for this but in more recent versions of SQLite the method below is used
DBInterface.execute(db,"SELECT * FROM movies")
```

```julia
SQLite.columns(db,"movies")
```
